```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import data
df = pd.read_csv('Europe Hotel Booking Satisfaction Score.csv')
df
```

```python
# Lihat sekilas data
print(df.head())
```

```python
print(df.info())
```

```python
# Tampilkan nama kolom untuk memastikan kolom yang tersedia
print(df.columns)
```

```python
print(df.describe())
```

Distribusi kepuasan pelanggan hotel

```python
df['satisfaction'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribusi Kepuasan Pelanggan Hotel')
plt.ylabel('')
plt.show()
```

```python
sns.countplot(x='Gender', hue='satisfaction', data=df)
plt.title('Distribusi Pelanggan Berdasarkan Gender dan Kepuasan')
plt.xlabel('Gender')
plt.ylabel('Jumlah')
plt.show()
```

```python
df['Gender'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Pelanggan Hotel berdasarkan Gender')
plt.ylabel('')
plt.show()
```

```python
sns.histplot(data=df, x='Age', hue='satisfaction', kde=True)
plt.title('Distribusi Kepuasan Berdasarkan Usia')
plt.xlabel('Usia')
plt.ylabel('Frekuensi')
plt.show()
```
